# Linear Regression

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[1], [2], [3]])
```

```python
W = torch.rand(1, requires_grad=True) # set model
b = torch.rand(1, requires_grad=True)
```

```python
def criterion(y_hat, y): # set loss function
  return torch.mean((y_hat - y) ** 2)
```

```python
optimizer = optim.SGD([W, b], lr=0.01) # set optimizer
```

```python
epochs = 100
for epoch in range(epochs):
  hypothesis = x_train * W + b # forward propagation
  cost = criterion(hypothesis, y_train) # get cost

  optimizer.zero_grad() # reset previous computed gradient value to 0
  cost.backward() # backward propagation
  optimizer.step() # update parameters

  print('Epoch {:4d}/{} Cost: {:.6f} W {:.3f} b: {:.3f}'.format(\
      epoch, epochs, cost.item(), W.item(), b.item()))
```

```python
def criterion(y_hat, y):
  return torch.mean(torch.square(y_hat - y))
```

```python
class LinearRegression(nn.Module): # nn.Module을 상속받기
  def __init__(self, x_in, x_out): # x_in, x_out은 각각 input과 output의 dimension
    super(LinearRegression, self).__init__()
    self.linear = nn.Linear(x_in, x_out)
  def forward(self, x):
    return self.linear(x)
```

```python
model = LinearRegression(1, 1).train() # Q) 여기선 Logistic Regression이랑 다르게 train()을 왜 안붙혀? 실수...
optimizer = optim.SGD(model.parameters(), lr = 0.01)
```

```python
epochs = 30
for epoch in range(epochs):
  # model.train() 이걸 왜 안넣어줘? 실수...
  model.train()
  hypothesis = model.forward(x_train) # forward propagation
  # model(x_train) 이건 model.forward를 부르는 것과 같다
  cost = criterion(hypothesis, y_train) # get cost

  optimizer.zero_grad()
  cost.backward() # backward propagation
  optimizer.step() # update parameters

  print('Epoch {:4d}/{} Cost: {:.6f} W {:.3f} b: {:.3f}'.format(\
  epoch, epochs, cost.item(), model.linear.weight.item(), model.linear.bias.item()))
```

# Logistic Regression

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x_train = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = torch.FloatTensor([[0], [0], [0], [1]])
```

```python
class LogisticRegression(nn.Module):
  def __init__(self, x_in, x_out):
    super(LogisticRegression, self).__init__()
    self.linear = nn.Linear(x_in, x_out)
    self.activation = nn.Sigmoid()
  def forward(self, x):
    z = self.linear(x)
    a = self.activation(z)
    return a
```

```python
model = LogisticRegression(2, 1).train() # Q) train을 붙히고 말고는 어떻게 판단? 붙혀야해..... 안붙혀도 디폴트로들어가긴 해
```

```python
optimizer = optim.SGD(model.parameters(), lr=0.01) # set optimizer
```

```python
criterion = nn.BCELoss()
```

```python
epochs = 1000
for epoch in range(epochs):
  model.train()
  hypothesis = model.forward(x_train) # forward propagation
  cost = criterion(hypothesis, y_train) # Q) hypothesis에 1e-8이란 작은값을 더해주는 이유는? 1e-8 안붙혀도 돼

  optimizer.zero_grad()
  cost.backward() # backward propagation
  optimizer.step() # update parameters

  if epoch != 0 and epoch % 100 == 0:
    model.eval()
    with torch.no_grad():
        predicts = (model.forward(x_train))
        print('predict with model : {}'.format(predicts))
        print('real value y : {}'.format(y_train))
```

# Multi-Layer Perceptron

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x_train = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = torch.FloatTensor([[0], [1], [1], [0]])
```

```python
class MultiLayerPerceptron(nn.Module):
  def __init__(self):
    super(MultiLayerPerceptron, self).__init__()
    self.linear1 = nn.Linear(2, 3)
    self.activation = nn.Sigmoid()
    self.linear2 = nn.Linear(3, 1)

  def forward(self, x):
    z1 = self.linear1(x)
    a1 = self.activation(z1)

    z2 = self.linear2(a1)
    a2 = self.activation(z2)
    return a2
```

```python
model = MultiLayerPerceptron().train()
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 10000
for epoch in range(epochs):
  model.train()
  hypothesis = model.forward(x_train)
  cost = criterion(hypothesis, y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch != 0 and epoch % 100 == 0:
    model.eval()
    with torch.no_grad():
        predicts = (model.forward(x_train))
        print('predict with model : {}'.format(predicts))
        print('real value y : {}'.format(y_train))
```
